```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
def pre_processing(file_name):
    acce_result = []
    gyro_result = []
    with open(file_name, 'r') as input_file:
        for line in input_file.readlines():
            parsed = line.strip().split(' ')
            if parsed[0] == 'Gyro:':
                gyro_result.append([float(i) for i in parsed[1:]])
            elif parsed[0] == 'Acce:':
                acce_result.append([float(i) for i in parsed[1:]])
            else:
                print('cannot parse ' + line)
    return np.array(acce_result), np.array(gyro_result)

def plot_data(data, data_name):
    plt.plot(data[:, 0], label='x-axis')
    plt.plot(data[:, 1], label='y-axis')
    plt.plot(data[:, 2], label='z-axis')
    plt.xlabel('samples')
    plt.ylabel('value')
    plt.title("{} over time".format(data_name))
```

```python
acce_data, gyro_data = pre_processing('place_still_over_1_min.txt')
```

```python
plot_data(acce_data, 'accelerometer raw output')
plt.savefig('accelerometer_raw.png')
```

```python
plot_data(gyro_data, 'gyro raw output')
plt.savefig('gyro_raw.png')
```
